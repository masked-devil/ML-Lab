```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\91935\Downloads\Admission_Predict.csv')
```

```python
df["chance of admit 0/1"]=[0 if each<0.8 else 1 for each in df['Chance of Admit ']]
```

```python
df
```

```python
x=df[['GRE Score','TOEFL Score', 'SOP','LOR ','CGPA']]
y=df['chance of admit 0/1']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier(max_depth=2)
```

```python
model= clf.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
plt.figure(figsize=(8,8))
from sklearn.tree import plot_tree
a=plot_tree(model,filled=True)
```
